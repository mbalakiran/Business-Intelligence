```python
zomato_c = zomato.copy()
```

```python
zomato_c.info()
```

```python
del zomato_c['url']
del zomato_c['address']
del zomato_c['phone']
del zomato_c['location']
```

```python
zomato_c.rename(columns={'approx_cost(for two people)': 'average_cost', 'listed_in(city)': 'locality','listed_in(type)': 'restaurant_type'}, inplace=True)
```

```python
zomato_c.rate = zomato_c.rate.replace("NEW", np.nan)
zomato_c.dropna(how ='any', inplace = True)
```

```python
zomato_c.info()
```

```python
plt.rcParams['figure.figsize'] = (3, 4)
plt.style.use('_classic_test')

zomato_c['online_order'].value_counts().plot.bar(color = 'cyan')
plt.title('Online orders', fontsize = 20)
plt.ylabel('Number of orders', fontsize = 15)
plt.show()
```

```python
rcParams['figure.figsize'] = 15,7
g = sns.countplot(x="locality",data=zomato_c, palette = "Set1")
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
g 
plt.title('locality',size = 20)
```

```python
rcParams['figure.figsize'] = 15,7
g = sns.countplot(x="rest_type",data=zomato_c, palette = "Set1")
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
g 
plt.title('rest_type',size = 20)
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
x = pd.crosstab(zomato_c['rate'], zomato_c['online_order'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True,color=['red','yellow'])
plt.title('online order vs rate', fontweight = 30, fontsize = 20)
plt.legend(loc="upper right")
plt.show()
```

```python
zomato_c.head()
zomato_c.average_cost = zomato_c.average_cost.apply(lambda x: x.replace(',',''))
zomato_c.average_cost = zomato_c.average_cost.astype(int)
fig, ax = plt.subplots(figsize=[16,4])
sns.distplot(zomato_c['average_cost'],ax=ax)
ax.set_title('Cost Distrubution for all restaurants')
```

```python
restaurantTypeCount=zomato_c['restaurant_type'].value_counts().sort_values(ascending=True)
slices=[restaurantTypeCount[0],
        restaurantTypeCount[1],
        restaurantTypeCount[2],
        restaurantTypeCount[3],
        restaurantTypeCount[4],
        restaurantTypeCount[5],
        restaurantTypeCount[6]]
labels=['Pubs and bars','Buffet','Drinks & nightlife','Cafes','Desserts','Dine-out','Delivery ']
colors = ['#3333cc','#ffff1a','#ff3333','#c2c2d6','#6699ff','#c4ff4d','#339933']
plt.pie(slices,colors=colors, labels=labels, autopct='%1.0f%%', pctdistance=.5, labeldistance=1.2,shadow=True)
fig = plt.gcf()
plt.title("Percentage of Restaurants according to their Type", bbox={'facecolor':'2', 'pad':5})

fig.set_size_inches(12,12)
plt.show()
```
